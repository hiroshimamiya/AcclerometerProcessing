```python
all_epoch_dfs = []
for epochfilename, labelfilename in zip(epoch_files, label_files):
    epoch_df = pd.read_csv(epochfilename)
    label_df = pd.read_csv(labelfilename)
    
    # retrieve only the activities headers
    activities = label_df[label_df.columns.values.tolist()[2:6]]
    
    # whichever activity is depicted by 1, use it as the predicted activity
    out = activities[activities==1].idxmax(axis=1)
    label_df["activity_predicted"] = out
    
    epoch_df["participant"] = float(epochfilename.split("/")[-1].split("-")[0][1:])
    epoch_df["MET"] = label_df["MET"]
    epoch_df["label"] = label_df["activity_predicted"]
    all_epoch_dfs.append(epoch_df)
```

```python
# LOOCV: leave one out cross validation for 151 participants

from sklearn.model_selection import LeaveOneGroupOut
from imblearn.ensemble import BalancedRandomForestClassifier
import numpy as np

groups = df['participant'].values

def _Model(**kwargs):
    return BalancedRandomForestClassifier(
        n_estimators=1000,
        max_depth=10,
        min_samples_leaf=1,
        replacement=True,
        sampling_strategy='not minority',
        random_state=42,
        **kwargs
    )

# Initialize LeaveOneGroupOut
logo = LeaveOneGroupOut()

# Lists to store outcomes for analysis
predictions = []
true_labels = []
```

```python
feature_columns = [col for col in df.columns if col not in ('participant', 'MET', 'label')]
X = df[feature_columns]
y = df['label'].values  # Labels for training
```

```python
logo.split(X, y, groups)
```

```python
# Iterate over each training/test set defined by LeaveOneGroupOut
for train_index, test_index in logo.split(X, y, groups):
    # Split the data
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    # Initialize and train the model
    model = _Model(n_jobs=100, verbose=1)
    model.fit(X_train, y_train)
    
    # Predict on the test set
    y_pred = model.predict(X_test)
    
    # Store the predictions and true labels for further evaluation
    predictions.extend(y_pred)
    true_labels.extend(y_test)
```
